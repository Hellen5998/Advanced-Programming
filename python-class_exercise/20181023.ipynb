```python
from selenium import webdriver
```

```python
browser=webdriver.Chrome()
```

```python
browser.get('https://www.baidu.com/s?ie=utf-8&f=8&rsv_bp=0&rsv_idx=1&tn=baidu&wd=李响')
```

```python
browser.title
```

```python
dir(browser)
```

```python
browser.page_source    #显示网页html
```

```python
html=browser.page_source
```

```python
from pyquery import PyQuery as pq
```

```python
doc=pq(html)
```

```python
doc('#1')
```

```python
doc1=pq(doc('#1').html())     #pq括号中写html格式的变量
```

```python
doc1
```

```python
doc1.html()
```

```python
doc1('h3')
```

```python
doc1('h3 a')
```

```python
doc1('h3 a').attr('href')
```

```python
url=doc1('h3 a').attr('href')
```

```python
br=webdriver.Chrome()
```

```python
br.get(url)
```

```python
html1=br.page_source      #爬取整个页面的html
doc2=pq(html1)
doc2.text()
```
