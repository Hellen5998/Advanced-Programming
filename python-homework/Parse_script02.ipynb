```python
results = []#所有的文件
#删除空文件
files=os.listdir("D:\\Parse")
for file in files:
    if os.path.getsize(os.path.join("D:\\Parse", file)) == 0:  # 文件大小为0
        os.remove(os.path.join("D:\\Parse", file))  # 删除这个文件
    else:
        apath = os.path.join("D:\\Parse", file)#合并成一个完整路径
        results.append(apath)
    #print(results)
word=[]
for result in results:
    file_content=open(result, 'r',encoding='utf-8')
    for i in file_content.read():
        word.append(i)
```

```python
word
```

```python
import os
from pyquery import PyQuery as pq
from selenium import webdriver
browser=webdriver.Chrome()
browser.get('http://www.guoxuedashi.com/zidian/')
for i in word:
    input = browser.find_element_by_id('sokeyzi')
    input.clear()
    input.send_keys(i)
    input.send_keys(Keys.ENTER)
```
